Now let's get some data with historical datestamps and load that up. Once we have the data we can come up with ways to visualize the data. 

```python
historical_covid_data = pd.read_csv('https://covidtracking.com/api/v1/states/daily.csv')
historical_covid_data.head(10)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
covid_data['state']
plot = covid_data.sort_values(by='death', ascending=False).plot.bar(y='death', x='state', figsize=(15, 5))
```

Next, let's see if we can plot the historical data for AZ. First let's see if we can just get the data for AZ from the dataset.

```python
plot_data = historical_covid_data.loc[historical_covid_data['state']=='AZ']
```

```python
plot_data
```

```python
plot_data.plot(x='date', y='total')
```

```python
plot_data.dtypes
```

The Dataset has an integer date that needs to be fixed to allow oorrect plotting. Let's see if we can fix that with a lambda function.

```python
plot_data['DateTime'] = plot_data['date'].apply(lambda x: pd.to_datetime(str(x), format='%Y%m%d'))
```

```python
plot_data.head()
```

Success! Now we can try the plot again, using "DateTime" column for x.

```python
plot_data.plot(x='DateTime', y='total')
```

Let's make a few changes to our simple plot so it is easier to interpret.

```python
import seaborn as sns
sns.set(font_scale=1.4)
plot_data.plot(x='DateTime', y='total',figsize=(12, 10), linewidth=2.5, color='maroon')
plt.xlabel("Date", labelpad=15)
plt.ylabel("Total Cases", labelpad=15)
plt.title("Total Covid-19 Cases Identified: AZ", y=1.02, fontsize=22);
```

```python
import seaborn as sns
sns.set(font_scale=1.4)
plot_data.plot(x='DateTime', y='positiveIncrease',figsize=(12, 10), linewidth=2.5, color='blue')
plt.xlabel("Date", labelpad=15)
plt.ylabel("Total Cases", labelpad=15)
plt.title("Daily Covid-19 Cases Identified: AZ", y=1.02, fontsize=22);
```
